# (2)Implement all mutation and crossover operators available for integer representation and floating point representation.

## Binary Representaiton
* Mutation: Bitwise
* Recombination: one point crossover, n-point crossover, uniform crossover 

### Integer Representation
* Mutation: Random resetting, Creep
* Recombination: one point, n-point, uniform

### Real valued or floting point Representation
* Mutation: Uniform, non-uniform, self adaptive, uncorrelated, correlated
* Recombination: Simple, single, whole, blend

### Tree Representation
* Mutation: Replacing subtree with random tree
* Recombination: subtree crossover

```python
# Float representation
import random
import pandas as pd
def fitness_func(population):
    
    fitness=[]
    for pop_row in population:
        fit=0
        fit=pop_row[0]*1-pop_row[1]*2+pop_row[2]*3-pop_row[3]*4 + pop_row[4]*6 # 1x-2y+3z-4a+6c
        
        fitness.append(fit)
    fitness =  [abs(ele) for ele in fitness] 
    print('The best population',population[fitness.index(min(fitness))])
    print('Best fitness value',min(fitness))
    return fitness

    
def select_parents(population, fitness):
    
    parents = []
    parent_ids=[]
#     fitness=abs(fitness)
    
    min_fitness_idx = numpy.where(fitness == numpy.min(fitness))[0][0]
    population=pd.DataFrame(population)
    parent_ids.append(min_fitness_idx)
    parents.append(list(population.iloc[min_fitness_idx,:]))
    
    fitness[min_fitness_idx]=9999
    
    min_fitness_idx = numpy.where(fitness == numpy.min(fitness))[0][0]
    parents.append(list(population.iloc[min_fitness_idx,:]))
    parent_ids.append(min_fitness_idx)
    
    return parents,parent_ids


def crossover(parents):
    
    #1-point crossover
    size=len(parents[0])
    n=random.randint(1,size-1)
    parents=pd.DataFrame(parents)

    cut1=parents.iloc[0,n:].copy()
    cut2=parents.iloc[1,n:].copy()
    
    parents.iloc[0,n:]=cut2
    parents.iloc[1,n:]=cut1
    

    return parents.values

def n_crossover(parents,n):
    print(parents[0])
    print(parents[1])
        
    #n-point crossover
    size=len(parents[0])
    n_idx=(random.sample(range(size), n))
    n_idx.sort()
    parents=pd.DataFrame(parents)
    print("\n N-points")
    print(n_idx)
    i=0
    while i<n:
        start=n_idx[i]
        
        if(i+1<n):
            end=n_idx[i+1]
            
        else:
            end=size-1
            
        cut1=parents.iloc[0,start:end].copy()
        cut2=parents.iloc[1,start:end].copy()

        parents.iloc[0,start:end]=cut2
        parents.iloc[1,start:end]=cut1
        
        i=i+2
    
    print("\n New Parents")
    print(parents.values)
    
    return parents.values

def mutation(parents):
    new_parent=[]
    for parent_row in parents:
        idx1=random.randint(0,len(parents[0])-1)
        idx2=random.randint(0,len(parents[0])-1)
        parent_row[idx1]=random.uniform(0,30)
        parent_row[idx2]=random.uniform(0,30)
        new_parent.append(list(parent_row))
    
    return list(new_parent)

def uniform_mutation(p1,p2):
        bit = random.randint(0,1)
        if(bit):
            p1[bit] = np.min(np.array([p1,p1])) + random.random()*( np.max(np.array([p1,p1])) - np.min(np.array([p1,p1])) )
            p2[bit] = np.min(np.array([p1,p1])) + random.random()*( np.max(np.array([p1,p1])) - np.min(np.array([p1,p1])) )
        else:
            p1[bit] = np.min(np.array([p1,p1])) + random.random()*( np.max(np.array([p1,p1])) - np.min(np.array([p1,p1])) )
            p2[bit] = np.min(np.array([p1,p1])) + random.random()*( np.max(np.array([p1,p1])) - np.min(np.array([p1,p1])) )
        return p1,p2
```
